```python
from google.colab import output
output.enable_custom_widget_manager()
import gmaps

gmaps.configure(api_key="hidden")

location = (34.054843, -84.174)
zoom_level = 12

# Create the map
fig = gmaps.figure(center=location, zoom_level=zoom_level)

# Add a layer for the traffic data
traffic_layer = gmaps.traffic_layer()
fig.add_layer(traffic_layer)

# Display the map
fig
```

```python
import folium

# Based on gmaps API
traffic_data = {
    'Intersection1': {'lat': 34.054843, 'lon': -84.174, 'traffic_level': 2},
    'Intersection2': {'lat': 34.05016048855818, 'lon': -84.17587310468359, 'traffic_level': 3},
    'Intersection3': {'lat': 34.04780349448761, 'lon': -84.17742752577927, 'traffic_level': 1},
    'Intersection4': {'lat': 34.036077792504926, 'lon': -84.17343246139271, 'traffic_level': 1},
}

m = folium.Map(location=[34.054843, -84.174], zoom_start=13)

for intersection, data in traffic_data.items():
    color = 'red' if data['traffic_level'] == 3 else 'orange' if data['traffic_level'] == 2 else 'green'
    folium.Marker(location=[data['lat'], data['lon']], popup=intersection, icon=folium.Icon(color=color)).add_to(m)

m.save('traffic_map.html')

# Show Map
m
```
